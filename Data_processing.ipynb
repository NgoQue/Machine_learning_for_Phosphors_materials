```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Data_final.csv", encoding="ISO-8859-1")
data = data.drop(columns=["Unnamed: 6", "Unnamed: 7", "Unnamed: 8"], errors="ignore")
data.head()
```

```python
element_symbols = [
    "H", "He", "Li", "Be", "B", "C", "N", "O", "F", "Ne", "Na", "Mg", "Al", "Si", "P", "S", "Cl", "Ar",
    "K", "Ca", "Sc", "Ti", "V", "Cr", "Mn", "Fe", "Co", "Ni", "Cu", "Zn", "Ga", "Ge", "As", "Se", "Br",
    "Kr", "Rb", "Sr", "Y", "Zr", "Nb", "Mo", "Tc", "Ru", "Rh", "Pd", "Ag", "Cd", "In", "Sn", "Sb", "Te",
    "I", "Xe", "Cs", "Ba", "La", "Ce", "Pr", "Nd", "Pm", "Sm", "Eu", "Gd", "Tb", "Dy", "Ho", "Er", "Tm",
    "Yb", "Lu", "Hf", "Ta", "W", "Re", "Os", "Ir", "Pt", "Au", "Hg", "Tl", "Pb", "Bi", "Po", "At", "Rn",
    "Fr", "Ra", "Ac", "Th", "Pa", "U", "Np", "Pu", "Am", "Cm", "Bk", "Cf", "Es", "Fm", "Md", "No", "Lr",
    "Rf", "Db", "Sg", "Bh", "Hs"]

# Add new cols in DataFrame
for column_name in element_symbols:
    data[column_name] = 0  # Add a new column with the default name and value of None. 

# Step 3: Save the results back to the CSV file.
output_file_path = 'updated_file_with_new_columns.csv'  
data.to_csv(output_file_path, index=False)
data.head()
```

```python
composition = data['Formula']
composition 
```

```python
np.shape(composition)
composition[1]
```

```python
import re
i = 0
for comp in composition:
    materials = re.findall(r'([A-Z][a-z]*)([\d\.]*)', comp)
    for material in materials:
        if material[0] in element_symbols:
            # print(material[0])
            fraction = float(material[1]) if material[1] else 1.0  
            data.at[i, material[0]] = fraction

    i += 1
output_file_path = 'updated_file_with_new_columns.csv' 
data.to_csv(output_file_path, index=False)
data.head()
```
