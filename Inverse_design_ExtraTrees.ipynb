```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Data_processed.csv")
data.head()
```

```python
data.columns
```

```python
data.info()
```

```python
# features = ['Emission']
features = ['Excitation', 'Emission']
a = data[features]
a = a.values
np.shape(a), a
```

```python
results  = ['H', 'Li', 'Be', 'B', 'C', 'N', 'O', 'F',
       'Na', 'Mg', 'Al', 'Si', 'P', 'Cl', 'K', 'Ca', 'Sc', 'Ti', 'V', 'Mn',
       'Co', 'Zn', 'Ga', 'Ge', 'Rb', 'Sr', 'Y', 'Zr', 'Nb', 'Mo', 'In', 'Sn',
       'Sb', 'Te', 'Cs', 'Ba', 'La', 'Eu', 'Gd', 'Lu', 'Ta', 'W', 'Pb', 'Bi']
b = data[results]
b = b.values
np.shape(b), b
```

### Spliting dataset

```python
from sklearn.model_selection import train_test_split
a_train, a_valid, b_train, b_valid = train_test_split(a, b, test_size=0.2, random_state = 24)
```

```python
a_train.shape, b_valid.shape
```

### 1.ExtraTreesRegressor

```python
from sklearn.ensemble import ExtraTreesRegressor
et_model = ExtraTreesRegressor(random_state =1)
et_model.fit(a_train, b_train)
et_preds = et_model.predict(a_valid)
et_preds.shape
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print("RMSR ExtraTreesRegressor: %f" %np.sqrt(mean_squared_error(b_valid, et_preds)))
print("R2 ExtraTreesRegressor: %f %%" %(100*r2_score(b_valid, et_preds)))
```

```python
et_model.predict([[350, 680]])
```

```python
from joblib import dump

dump(et_model, "extratrees_inverse_model.joblib")
```

```python
import pandas as pd
import numpy as np

if len(et_preds.shape) == 2 and et_preds.shape[1] == len(results):
    df = pd.DataFrame(et_preds, columns=results)
else:
    # Nếu chỉ có 1 giá trị dự đoán mỗi mẫu
    df = pd.DataFrame({'Predicted_Value': et_preds})

# Thêm các cột đầu vào Excitation và Emission
df[features] = a_valid

# Đưa Excitation và Emission lên đầu
cols = features + [c for c in df.columns if c not in features]
df = df[cols]

# Lưu ra file CSV
df.to_csv("ET_predictions_inverse_design.csv", index=False)

print("✅ Successfull save Et_predictions_inverse_design.csv")

df.head()
```

```python
import pandas as pd

f_preds = []
f_trues = []

for i in range(len(b_valid)):
    f_pred = ''
    f_true = ''
    
    for j in range(len(results)):
        if et_preds[i, j] != 0:
            f_pred += f"{results[j]}{et_preds[i, j]:.2f}"  # Làm tròn 2 chữ số sau dấu phẩy
    f_preds.append(f_pred)

    for j in range(len(results)):
        if b_valid[i, j] != 0:
            f_true += f"{results[j]}{b_valid[i, j]:.2f}"  # Làm tròn 2 chữ số sau dấu phẩy
    f_trues.append(f_true)

# Tạo DataFrame từ a_valid (dữ liệu đặc trưng) và thêm cột công thức
data_preds = pd.DataFrame(a_valid, columns=features)
data_preds['Actual formula'] = f_trues
data_preds['Predicted formula'] = f_preds

# Ghi ra file CSV với 2 chữ số sau dấu phẩy cho các giá trị số
data_preds.to_csv('Inverse_design_Mn4+.csv', index=False, float_format="%.2f")

data_preds.head()
```
